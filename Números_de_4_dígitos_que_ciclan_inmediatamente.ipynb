```python
import numpy as np

n = 10000
x1, x2 = list(range(n)), []
for i in x1:
  x = i
  y = str(x**2)
  while len(y) < 8:
    y = "0" + y
  x2.append(int(y[2:-2]))

print(x1)
print(x2)

eq = list(np.array(x1)==np.array(x2))

i = 0
ciclo_ll = [[0,0]]
while i < n and True in eq[i+1:]:
  i = eq.index(True, i + 1)
  ciclo_ll = np.append(ciclo_ll,[[x1[i], x2[i]]] , axis = 0)

print(ciclo_ll)
```
